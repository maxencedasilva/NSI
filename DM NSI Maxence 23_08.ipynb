# Code avec la méthode du cours

```python
def bin_to_deci(nb_bin):
    n=str(nb_bin)
    puiss=len(n)-1
    nb_dec=0
    while puiss>=0:
        for i in n:
            nb_dec+=int(i)*2**puiss
            puiss-=1
    return nb_dec

def deci_to_bin(nb_deci):
    n=nb_deci
    nb_bin=str()
    while n!=0:
        nb_bin=str(n%2)+nb_bin
        n//=2
    nb_bin=int(nb_bin)
    return nb_bin

def saisie():
    sens=int(input("Conversion de déci vers bin(entrer 0) ou l'inverse(entrer 1) : "))
    if sens==0:
        try:
            nb=int(input("Saisir le nombre à convertir en binaire : "))
            for i in str(nb):
                assert 9>=int(i)>=0
        except:
            print("Votre saisie est invalide.")
        else:
            print(f"Le nombre binaire de {nb} est {deci_to_bin(nb)}.")
    if sens==1:
        try:
            nb=int(input("Saisir le nombre à convertir en décimal : "))
            for i in str(nb):
                assert 1>=int(i)>=0
        except:
            print("Votre saisie est invalide.")
        else:
            print(f"Le nombre binaire de {nb} est {bin_to_deci(nb)}.")  
```

# Code sans la méthode du cours

```python
def bin_to_deci(nb_bin):
    n=int(str(nb_bin),2)
    return n

def deci_to_bin(nb_deci):
    n=bin(nb_deci)
    m=int(n[2:])
    return m

def saisie():
    sens=int(input("Conversion de déci vers bin(entrer 0) ou l'inverse(entrer 1) : "))
    if sens==0:
        try:
            nb=int(input("Saisir le nombre à convertir en binaire : "))
            for i in str(nb):
                assert 9>=int(i)>=0
        except:
            print("Votre saisie est invalide.")
        else:
            print(f"Le nombre binaire de {nb} est {deci_to_bin(nb)}.")
    if sens==1:
        try:
            nb=int(input("Saisir le nombre à convertir en décimal : "))
            for i in str(nb):
                assert 1>=int(i)>=0
        except:
            print("Votre saisie est invalide.")
        else:
            print(f"Le nombre binaire de {nb} est {bin_to_deci(nb)}.")
```

# Binaire vers hexadécimal et inversement

```python
def bin_to_hexa(nb_bin):
    n=hex(int(str(nb_bin),2))
    m=n[2:]
    return m

def hexa_to_bin(nb_hexa):
    n=bin(int(str(nb_hexa),16))
    m=n[2:]
    return m
```

# Conversion vers une base choisie

```python
symboles="0123456789ABCDEF" ## On limite ici à 16 symboles (base 16)
def base_vers_base(x,b1,b2):
    m=int(str(x),b1)
    if not x:
        return symboles[0]
    else:
        n=""
        while x!=0:
            n=symboles[m%b2]+n
            m//=b2
        return n ## On rend un str
```

## - Autre méthode avec un appel récursif

```python
symboles="0123456789ABCDEF" ## On limite ici à 16 symboles (base 16)
def base_vers_base(x,b1,b2):
    n=int(str(x),b1)
    if not n:
        return "" ## Renvoie une chaîne vide si x = 0 (ce qui évite l'affichage d'un zéro en trop)
    else:
        n=base_vers_base(n//b2,b1,b2)+symboles[n%b2]
        return n
```
