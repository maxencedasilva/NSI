## Exemple 5

```python
def insert_etoile(ch1):
    ch2=ch1[0]
    for i in range(1,len(ch1)):
        ch2=ch2+'*'+ch1[i]
    return ch2

ch=input('Saisir la chaîne de caractères: ')
print(insert_etoile(ch))
```

## Exemple 6

```python
def tiret(ch):
    ch2=''
    for k in ch:
        if k==' ':
            ch2=ch2+'-'
        else:
            ch2=ch2+k
    return ch2

ch=input('Saisir la chaîne de caractères: ')
print(tiret(ch))
```

## Exemple 8

```python
def inversement(ch):
    ch1=ch[0:3]
    ch2=ch[3:6]
    ch3=ch[6:9]
    return ch3+ch2+ch1

ch=input('Saisir la chaîne de 9 caractères: ')
print(inversement(ch))
```

## Exemple 12

```python
def espace(ch):
    ch2=''
    for k in ch:
        if k!=' ':
            ch2=ch2+k
    return ch2

def inverse(ch):
    ch_inv=ch[-1::-1]
    return ch_inv

ch=input('Saisir la chaîne sans accents, sans virgules et sans apostrophes: ')
ch2=inverse(espace(ch))
if espace(ch)==ch2:
    print(ch,'est un palindrôme.')
else:
    print(ch,"n'est pas un palindrôme.")
```
