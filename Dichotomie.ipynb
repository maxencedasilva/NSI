# Exercice 4

```python
from random import randint

def dicho(x,L):
    L.sort()
    g=0
    d=len(L)-1
    trouv=0
    while trouv==0 and g<=d:
        m=(g+d)//2
        if L[m]==x:
            trouv=1
        elif L[m]<x:
            g=m+1
        else:
            d=m-1
    if trouv==0:
        print('Cible non trouvée.')
    else:
        print("Cible trouvée à l'indice %s." %(m))

L=[randint(1,10) for _ in range(10)]
x=int(input('Donner la cible : '))
dicho(x,L)
```
