## Exemple 2

```python
from random import randint

def renvoie_ligne(M,l):
    return M[l]

def renvoie_colonne(M,c):
    return M[:][c]

def affiche_en_colonne(L):
    col=''
    for i in range(len(L)):
        col+='['+str(L[i])+']'+'\n'
    return col

M=[[randint(0,1) for a in range(5)] for b in range(3)]
affiche_en_colonne(renvoie_colonne(M,1))
```
